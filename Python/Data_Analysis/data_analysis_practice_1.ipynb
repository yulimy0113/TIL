### Question 14
각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라.

```python
df1 = channel[['channelname', 'subcnt', 'ct']][channel['ct'] > '2021-10-03'].sort_values('ct').drop_duplicates('channelname')
df2 = channel[['channelname', 'subcnt', 'ct']][(channel['ct'] > '2021-10-03 03:00:00') & (channel['ct'] < '2021-11-01 15:00:00')].sort_values('ct').drop_duplicates('channelname', keep = 'last')
```

```python
df1 = df1[['channelname', 'subcnt']].sort_values('channelname').reset_index(drop = True)
df2 = df2[['channelname', 'subcnt']].sort_values('channelname').reset_index(drop = True)
```

```python
subcnt_df = df1.merge(df2, left_on='channelname', right_on='channelname')
subcnt_df['diff'] = subcnt_df['subcnt_y'] - subcnt_df['subcnt_x']
```

```python
answer = subcnt_df[['channelname', 'diff']]
answer
```

### Question 15
각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다.  
공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분이상인 데이터 구간( 해당 시점 전,후) 의 시각을 모두 출력하라.

```python
import datetime
ep_1 = video[video['videoname'] == ' 공범 EP1'].reset_index(drop = True)
```

```python
less_5 = ep_1[ep_1.ct.diff(1) <= datetime.timedelta(minutes=5)]
longer_20 = ep_1[ep_1.ct.diff(1) >= datetime.timedelta(minutes=20)]
```

```python
answer = ep_1.loc[list(less_5.index) + list(less_5.index -1) + list(longer_20.index) + list(longer_20.index -1), :].sort_index()
answer
```

### Question 17
“공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.  
공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라.

```python
released_date = video[pd.to_datetime(video['ct']).dt.hour == 21]
```

### Question 18
video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고  
videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라.

```python
ratio_df = video.sort_values('ct', ascending = False).drop_duplicates('videoname')
```

```python
ratio_df['ratio'] = ratio_df['dislikecnt']/ratio_df['likecnt']
```

```python
answer = ratio_df[['videoname', 'ratio']].sort_values('ratio').reset_index(drop = True)
answer
```

### Question 19
2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오.

```python
df = video[(video['ct'] >= '2021-11-01 00:00:00') & (video['ct'] <= '2021-11-01 15:00:00')].sort_values('ct')
check_diff = lambda viewcnt: max(viewcnt) - min(viewcnt)
```

```python
answer = df[['videoname', 'viewcnt']].groupby('videoname', as_index = False).agg(check_diff)
answer
```
